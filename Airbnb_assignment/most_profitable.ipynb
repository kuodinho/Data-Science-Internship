```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('listings_am.csv')
df['price_as_int']=df.price.str.replace('[\$\,\.]', '', regex=True).astype(int)/100
df.loc[:, 'day']=365-df['availability_365']
df.loc[:, 'max_profit']=df['day']*df['price_as_int']
available=df[df['has_availability']=='t'][['neighbourhood_cleansed','price','day','availability_365',"room_type",'has_availability','max_profit','number_of_reviews']]
busy_places=available.sort_values(by=['max_profit'],ascending=False)[['neighbourhood_cleansed','price','day','availability_365',"room_type",'has_availability','max_profit','number_of_reviews']].head(2100)
places_types=busy_places.groupby('room_type')['max_profit'].median().reset_index()
bplot=sns.catplot(x='room_type', y="max_profit", data=places_types, kind='bar')
bplot.set(xlabel="Most profitable room types in Amsterdam", ylabel="Median of maximum profit")
ax = bplot.axes.flat[0]
# iterate through the axes containers
for c in ax.containers:
    labels = [f'{(v.get_height() / 1000):.1f}K' for v in c]
    ax.bar_label(c, labels=labels, label_type='edge')
bplot.fig.suptitle("Best 30 % by performance in the next year in Amsterdam")
plt.show()
```
