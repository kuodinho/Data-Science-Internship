```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_amsterdam = pd.read_csv('listings_am.csv')
df_amsterdam['price_as_int']=df_amsterdam.price.str.replace('[\$\,\.]', '', regex=True).astype(int)/100
available=df_amsterdam[(df_amsterdam['availability_365']<60)&(df_amsterdam['has_availability']=='t')][['listing_url','neighbourhood_cleansed','property_type','price_as_int','price','availability_365',"room_type",'has_availability','number_of_reviews']]
df_amsterdam[['neighbourhood_cleansed','property_type','price','availability_365',"room_type",'has_availability','number_of_reviews']]
neighbourhoods=available.sort_values(by=['availability_365'],ascending=False)[['neighbourhood_cleansed','has_availability','listing_url','property_type','price_as_int','price','number_of_reviews','neighbourhood_cleansed', 'availability_365']]
neighbourhoods_counts=available[["neighbourhood_cleansed",'availability_365']].value_counts()
#neighbourhoods_by_type=available.groupby(["neighbourhood_cleansed","property_type"])["price_as_int"].mean()
```

```python
ax = neighbourhoods_counts.head(15).plot(kind='pie')
ax.set_title("15 most popular neighbourhoods from top booked types in Amsterdam")
```
